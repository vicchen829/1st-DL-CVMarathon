```python
import cv2
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
%matplotlib inline
```

```python
img=cv2.imread('lena.png')
print("opencv 是用 "+str(type(img))+" 表示圖片格式")
print("圖片維度: "+str(img.shape))
```

# Opencv在載入圖片時可選擇載入方式的格式 :
1. cv2.IMREAD_COLOR(預設)
    * 載入包含RGB 3個channel的彩色圖片
2. cv2.IMREAD_GRAYSCALE
    * 灰階格式
3. cv2.IMREAD_UNCHANGE
    * 所有channel

```python
cv2.imshow('rgb',img) #(顯示的視窗名稱,要顯示的圖片)
cv2.waitKey(0)
cv2.destroyAllWindows()
#後⾯面兩兩⾏行行代表顯⽰示圖片之後，按任意鍵關閉視窗
```

```python
img_g=cv2.imread('lena.png',cv2.IMREAD_GRAYSCALE)
```

```python
while True:
    cv2.imshow('rgb',img)
    cv2.imshow('gray',img_g)
    
    k=cv2.waitKey(0)
    if k == 27 :
        cv2.destroyAllWindows()
        break
```

```python
img_box=cv2.imread('box.png')
```

```python
#把顯示圖片包成一個程式
def show_img(window_name,img_name):
    cv2.imshow(window_name,img_name)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
show_img('box',img_box)
```

```python
#opencv 預設讀取圖片為bgr
img_b=img[:,:,0]
img_g=img[:,:,1]
img_r=img[:,:,2]
img_concat = np.hstack((img_b, img_g, img_r))
while True:
    cv2.imshow('img',img)
    cv2.imshow('img_concat',img_concat)
    k=cv2.waitKey(0)
    if k == 27 :
        cv2.destroyAllWindows()
        break
```

```python
#如果用plt讀取
img_plt=mpimg.imread('lena.png')
img_plt_r=img_plt[:,:,0]
img_plt_g=img_plt[:,:,1]
img_plt_b=img_plt[:,:,2]

plt.figure(num='lena',figsize=(8,8))

plt.subplot(2,2,1)
plt.title('original')
plt.imshow(img_plt)

plt.subplot(2,2,2)
plt.title('R channel')
plt.imshow(img_plt_r)

plt.subplot(2,2,3)
plt.title('G channel')
plt.imshow(img_plt_g)

plt.subplot(2,2,4)
plt.title('B channel')
plt.imshow(img_plt_b)
```
